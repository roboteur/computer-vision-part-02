## EYELEE v1.0 by The Roboteur

```
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Activation
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
```

```
print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU')))
```

```
resnet50_model = tf.keras.applications.ResNet50()
```

```
x = resnet50_model.layers[-2].output           
predictions = Dense(2, activation='softmax')(x) 
eyelee_model = Model(inputs=resnet50_model.input, outputs=predictions)
```

```
for layer in eyelee_model.layers [:-1]:
  layer.trainable = False
```

```
eyelee_model.compile(Adam(lr=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```
train_path= '../input/rockpaper/train'
valid_path= '../input/rockpaper/valid'
test_path= '../input/rockpaper/test'
```

```
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.resnet.preprocess_input).flow_from_directory(train_path, target_size=(224, 224), classes=['rock', 'paper'], batch_size=500)
valid_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.resnet.preprocess_input).flow_from_directory(valid_path, target_size=(224, 224), classes=['rock', 'paper'], batch_size=20)
test_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.resnet.preprocess_input).flow_from_directory(test_path, target_size=(224, 224), classes=['rock', 'paper'], batch_size=10)
```

##### III ] TRAIN AND SAVE THE MODEL

```
eyelee_model.compile(Adam(lr=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```
eyelee_model.fit_generator(train_batches, steps_per_epoch=4, validation_data=valid_batches, validation_steps=4, epochs=80, verbose=2)
```

```
# Save to the model to .h5
eyelee_model.save('eyelee_v1.h5')
```
